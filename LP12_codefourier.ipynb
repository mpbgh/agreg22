```python
import numpy as np
import matplotlib.pyplot as plt
```

Série de Fourier fonction carrée

```python
T = 1
nT = 2
t = np.linspace(0, T*nT, 1000)
f = 1/T

A0 = 1

ptot = 2
st = np.zeros(len(t))
for i in range (ptot):
    n = 2*i+1
    sh = 4 * A0 / (np.pi * n) * np.sin(2*np.pi*n*f*t)
    st += sh
    
    if i <= 3 : 
        plt.plot(t,sh, label = 'n = ' + str(n))
        
    
plt.legend(loc = 'best')
plt.xlabel('t')
plt.ylabel('harmonique n')
plt.show()

plt.plot(t,st)
plt.xlabel('t')
plt.ylabel('s(t)')
plt.show()
```

Série de Fourier fonction triangle

```python
T = 1
nT = 2
t = np.linspace(0, T*nT, 1000)
f = 1/T

A0 = 1

ptot = 4
st = np.zeros(len(t))
for i in range (ptot):
    n = 2*i+1
    sh = 4 * A0 / (np.pi * n)**2 * np.cos(2*np.pi*n*f*t)
    st += sh
    
    if i <= 3 : 
        plt.plot(t,sh, label = 'n = ' + str(n))
        
    
plt.legend(loc = 'best')
plt.xlabel('t')
plt.ylabel('harmonique n')
plt.show()

plt.plot(t,st)
plt.xlabel('t')
plt.ylabel('s(t)')
plt.show()
```
