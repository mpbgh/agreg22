```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
def xp_f(x, w0, C):
    return np.sqrt(2* w0**2 * np.cos(x) + C)
```

```python
def C_f(x0, w0, xp0):
    return xp0**2 - 2* w0**2 * np.cos(x0)
```

```python
matplotlib.rcParams['contour.negative_linestyle'] = 'solid'

l = 1 #m
g = 9.81

w0 = np.sqrt(g/l)

x1_min = -4.0
x1_max = 4.0
x2_min = -3.0
x2_max = 3.0

x1, x2 = np.meshgrid(np.arange(x1_min,x1_max, 0.01), np.arange(x2_min,x2_max, 0.01))

y = C_f(x1, w0, x2*w0)

Ccrit = 2*w0**2

fig = plt.figure(figsize=(10, 20))
ax = fig.add_subplot(111)
#plt.imshow(y,extent=[x1_min,x1_max,x2_min,x2_max])
#plt.contour(x1, x2, y, [-75,-50,-10,20,50,69,78.959,100,150,210,300,400], colors = 'k')
plt.contour(x1, x2, y, [-Ccrit/1.01,-Ccrit/1.1, -Ccrit/1.3, -Ccrit/1.7, -Ccrit/2.7, -Ccrit/10, Ccrit/4, Ccrit/1.5, Ccrit/1.1, Ccrit, Ccrit*1.2, Ccrit*2, Ccrit*4], colors = 'k')



plt.ylabel("$\dot{\\theta}/\omega_0$" , fontsize=16)
plt.xlabel("$\\theta$" , fontsize=16)


ax.set_aspect('equal')

```
